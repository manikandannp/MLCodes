```python
global offset
offset = '0'
limit = '100'
total = '10000'

AIRTABLE_BASE_ID = 'appcBVMfuKAt60RSi'
AIRTABLE_TABLE_NAME = 'Tanner\'s%20View'
API_KEY = 'key3lVkuEwSfK3N9n'
END_POINT = f'https://api.airtable.com/v0/{AIRTABLE_BASE_ID}/{AIRTABLE_TABLE_NAME}'
headers = {
    "Authorization": f"Bearer {API_KEY}"
#   "Content-Type": "application/json"
}
```

```python
result = []
while True :
    try :
        response = requests.get(END_POINT, headers = headers, params ={"pageSize": 100, "offset": offset, "total": total})
        response_Table = response.json()
        records = list(response_Table['records'])
        result.append(records)
    #   print('Page Num :' + str(len(result)))
    #   print(records[0]['id'] , len(records))
        try : 
            offset = response_Table['offset']
    #       print(offset)
        except Exception as ex:
    #       print(ex , offset)
            break
    except error as e:
        print(e)
print('Total Num of Pages Extracted: ' + str(len(result)))
#print(result)
```

```python
cols = ['ID', 'PLANNING_STATUS', 'LANDING_MARKET', 'TARGET_GEO' \
                , 'EVENT_TYPE', 'EVENT_NAME', 'START_DATE', 'CITY_AND_STATE' \
                , 'TOTAL_COST', 'OBJECTIVE', 'AUDIENCE_ROLE', 'ORIGINATING_TEAM' \
                , 'CAMPAIGN_CODE', 'END_DATE', 'REGION_MARKET', 'HUB_US_ONLY', 'EXPECTED_SIZE'
                , 'EVNT_BRIEF_ID', 'CREATETIME', 'DELIVERY_METHOD', 'EXPECTED_RESPONSES', 'LEVERS', 'QUARTER']
```

```python
df = pd.DataFrame(columns = cols)

def dict_key(str1, d):
    if str1 in d.keys():
    #    print(str1)
    #    print(isinstance(d[str1], list))
        if isinstance(d[str1], list) == True:
            str2 = ', '.join(map(str, d[str1]))
            return str2
        else:
            return d[str1]
    else: return '' #return 'NAN'

j = 0
while j < len(result):    
    for i in result[j]: #r_json["records"]
        id = dict_key('id', i)
        ps = dict_key('Planning Status', i['fields'])
        lm = dict_key('Landing Market', i['fields'])
        tg = dict_key('Target Geo', i['fields'])
        et = dict_key('Event Type', i['fields'])
        en = dict_key('Event Name', i['fields'])
        sd = dict_key('Start Date', i['fields'])
        cs = dict_key('City and State', i['fields'])
        tc = dict_key('Total Cost', i['fields'])
        ob = dict_key('Objective', i['fields'])
        ar = dict_key('Audience Role', i['fields'])
        ot = dict_key('Originating Team', i['fields'])
        cc = dict_key('Campaign Code', i['fields'])
        ed = dict_key('End Date', i['fields'])
        rm = dict_key('Americas Region and/or Market', i['fields'])
        hu = dict_key('Hub (US only)', i['fields'])
        es = dict_key('Expected Size', i['fields'])
        eb = dict_key('2022 EMB #', i['fields'])
        cr = dict_key('createdTime', i)
        qr = dict_key('Quarter', i['fields'])
        lv = dict_key('Levers', i['fields'])
        dm = dict_key('Delivery Method', i['fields'])
        er = dict_key('Expected Responses', i['fields'])
        df = df.append({'ID' : id, 'PLANNING_STATUS' : ps, 'LANDING_MARKET': lm, 
                'TARGET_GEO' : tg, 'EVENT_TYPE' : et, 'EVENT_NAME' : en,
                'START_DATE' : sd, 'CITY_AND_STATE' : cs, 'TOTAL_COST' : tc,
                'OBJECTIVE' : ob, 'AUDIENCE_ROLE' : ar, 'ORIGINATING_TEAM' : ot,
                'CAMPAIGN_CODE' : cc, 'END_DATE' : ed, 'REGION_MARKET' : rm,
                'HUB_US_ONLY' : hu, 'EXPECTED_SIZE' : es, 'EVNT_BRIEF_ID' : eb,
                'CREATETIME' : cr,
                'QUARTER': qr, 'LEVERS': lv,
                'DELIVERY_METHOD': dm, 'EXPECTED_RESPONSES': er
                }, ignore_index = True)
    j = j + 1
print('Total Num of rows extracted: ' + str(len(df)))
```
